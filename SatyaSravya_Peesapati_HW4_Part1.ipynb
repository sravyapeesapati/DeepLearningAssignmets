```python
import cv2
import os

# Create a directory for frames
if not os.path.exists('frames'):
    os.mkdir('frames')

# Capture video
vidcap = cv2.VideoCapture('/content/Assignment.mp4')
success,image = vidcap.read()
count = 0

while success:
    cv2.imwrite("frames/frame%d.jpg" % count, image)     # save frame as JPEG file
    success,image = vidcap.read()
    print('Read a new frame: ', success)
    count += 1
```

```python
import os
import cv2
import torch

# Check current working directory
print("Current Working Directory:", os.getcwd())

# Assuming frames are in a directory named 'frames'
frames_dir = 'frames'
frame_files = [f for f in os.listdir(frames_dir) if f.endswith('.jpg')]

# Verify that frame files are found
print(f"Found {len(frame_files)} frame files.")

# Load the YOLOv5 model
model = torch.hub.load('ultralytics/yolov5', 'yolov5s', pretrained=True)

# Directory for saving processed frames
processed_frames_dir = 'processed_frames'
os.makedirs(processed_frames_dir, exist_ok=True)

# Process each frame file
for frame_file in sorted(frame_files):
    frame_path = os.path.join(frames_dir, frame_file)

    # Load image with OpenCV
    img = cv2.imread(frame_path)

    # Convert BGR image to RGB for model processing
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    # Inference
    results = model(img_rgb, size=640)

    # Retrieve the processed image (results.render() returns a list of images)
    processed_img_rgb = results.render()[0]

    # Convert RGB back to BGR to save with OpenCV
    processed_img_bgr = cv2.cvtColor(processed_img_rgb, cv2.COLOR_RGB2BGR)

    # Save processed image
    processed_frame_path = os.path.join(processed_frames_dir, frame_file)
    cv2.imwrite(processed_frame_path, processed_img_bgr)

    print(f'Processed and saved: {processed_frame_path}')
```

```python
import cv2
import os

frame_files = sorted(os.listdir('processed_frames'), key=lambda x: int(x.split('.')[0][5:]))  # Adjust the lambda function based on your filename structure

frame_array = []
out_path = 'vehicle_detection_output.mp4'
size = (0, 0)

for frame_file in frame_files:
    frame_path = f'processed_frames/{frame_file}'
    img = cv2.imread(frame_path)
    height, width, layers = img.shape
    size = (width, height)
    frame_array.append(img)

fps = 30

out = cv2.VideoWriter(out_path, cv2.VideoWriter_fourcc(*'DIVX'), fps, size)

for frame in frame_array:
    out.write(frame)
out.release()
```
